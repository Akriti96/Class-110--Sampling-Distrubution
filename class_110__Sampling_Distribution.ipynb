Lets take temperature data in a the room. We will draw a plot for a temperature data

```python
import plotly.figure_factory as ff
import plotly.graph_objects as go
import statistics
import random
import pandas as pd

df=pd.read_csv("data.csv")
# print(df.head())
data= df["temp"].tolist()
print(data)

#  cal mean
Mean= statistics.mean(data);
Std= statistics.stdev(data);
print("Mean: ",Mean);
print("Std: ",Std)

graph=ff.create_distplot([data],["Temp"],show_hist=False)
graph.show()
```

**Task: How to collect the 100 Random Data Points ?**



```python
import random
import statistics
dataset=[]
for i in range(0,100):
  random_index=random.randint(0,len(data))
  # print(random_index)
  value=data[random_index]
  dataset.append(value)
  # print(dataset)
mean= statistics.mean(dataset)
std_1= statistics.stdev(dataset)
print(mean)
print(std_1)
```

Lets do this 1000 times. Let us collect 1000 samples of 100 data points from the raw Data(Populatin Data). lets find the mean of each sample and store them in a list. then plot distribution of all samples on screen


```python
#  code to find the mean of 100 data points 1000 times and plot it
#  function to get the mean of given samples
# pass the number of data points you want as counter

def random_set_data(counter):
  dataset=[]
  for i in range(0,counter):
    random_index= random.randint(0,len(data)-1)
    value= data[random_index]
    dataset.append(value)
  mean= statistics.mean(dataset)
  return mean
    

#function to plot the mean on the graph
def showgraph(meanlist):
   df= meanlist
   mean= statistics.mean(df)
   g1= ff.create_distplot([df],["temp"],show_hist=False)
   g1.add_trace(go.Scatter(x=[mean,mean], y=[0,0.08], mode="lines", name="MEAN"))
   g1.show()
  
# function to get mean of 100 data Points 1000 times and plot the graph
def totaldata():
  meanlist=[]
  for i in range(0,1000):
    set_of_mean= random_set_data(100)
    meanlist.append(set_of_mean)
  showgraph(meanlist)
  mean= statistics.mean(meanlist)
  print(mean)

totaldata()
```

```python

#  code to find the standard deviation of the sample data
def std():
  meanlist=[]
  for i in range(0,1000):
    set_of_data= random_set_data(100)
    meanlist.append(set_of_data)

  std2= statistics.stdev(meanlist)
  print(std2)
std()

#  0.57
```
